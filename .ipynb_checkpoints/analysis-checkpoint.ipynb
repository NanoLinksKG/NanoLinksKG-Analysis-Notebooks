```python
from SPARQLWrapper import SPARQLWrapper, JSON
import json
from pandas import json_normalize

import pandas as pd
import numpy as np
import scipy
import math
import os

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline     
sns.set(color_codes=True)
%config InlineBackend.figure_format = 'retina'
```

```python
def query_endpoint(sparql_query, sparql_service_url):
    """
    Query the endpoint with the given query string and return the results as a pandas Dataframe.
    """
    sparql = SPARQLWrapper(sparql_service_url, agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36")  
    
    sparql.setQuery(sparql_query)
    sparql.setReturnFormat(JSON)

    # ask for the result
    result = sparql.query().convert()
    return json_normalize(result["results"]["bindings"])
```

```python
def plot_dose_response(nanomaterials):
    """
    Query the endpoint with the given query string and return the results as a pandas Dataframe.
    """
    sparql_query = """

    PREFIX bao: <http://www.bioassayontology.org/bao#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dct: <http://purl.org/dc/terms/>
    PREFIX enm: <http://purl.enanomapper.org/onto/>
    PREFIX npo: <http://purl.bioontology.org/ontology/npo#>
    PREFIX obo: <http://purl.obolibrary.org/obo/>
    PREFIX efo: <http://www.ebi.ac.uk/efo/>

    select distinct ?respValue ?concValue ?bioassay ?material ?bioassayLabel ?cellLineLabel ?organismLabel ?datasetTitle ?size ?mLabel where { 

        ?bioassay   rdfs:label ?bioassayLabel;
                    bao:BAO_0090012 ?material;
                    bao:BAO_0000205 ?assayFormat;
                    bao:BAO_0000209 ?mgSampleGroup1.

        ?assayFormat bao:BAO_0002004 ?cellLine.

        ?cellLine rdf:type ?cellLineType.
        
        ?cellLine bao:BAO_0090006 ?organism;
                  rdfs:label ?cellLineLabel.

        ?organism   a bao:BAO_0000551;
                    rdfs:label ?organismLabel.
        
        ?mgSampleGroup1 a bao:BAO_0000040;
                        bao:BAO_0000208 ?responseEp1 .
        
        ?responseEp1 bao:BAO_0000185 ?material;
                     bao:BAO_0000108 ?respValue;
                     bao:BAO_0002669 "24"^^xsd:float;
                     bao:BAO_0000338 ?concEp.

        VALUES ?mType {"""+nanomaterials+"""}
        ?material rdfs:label ?mLabel;
                  dct:type ?mType;
                  dct:source ?dataset;
                  npo:has_quality ?sizeEp.

        ?dataset rdfs:label ?datasetTitle.
        
        ?sizeEp a enm:ENM_8000039;
                bao:BAO_0095007 ?size.

        ?concEp a bao:BAO_0000098;
            bao:BAO_0002666 ?concValue.
        
        FILTER(?concValue >= 1 && ?concValue <= 100)

    } 
    """
    sparql_service_url = "http://virtuoso:8890/sparql"

    sparql = SPARQLWrapper(sparql_service_url, agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36")  
    
    sparql.setQuery(sparql_query)
    sparql.setReturnFormat(JSON)

    result = sparql.query().convert()
    result_table = json_normalize(result["results"]["bindings"])
    
    result_table = result_table.drop_duplicates()
    
    result_table['concValue.value'] = result_table['concValue.value'].astype(float)
    result_table['respValue.value'] = result_table['respValue.value'].astype(float)
    result_table['size.value'] = round(result_table['size.value'].astype(float),2)

    result_table = result_table[['concValue.value', 'respValue.value', 'material.value', 'bioassay.value', 
                            'mLabel.value', 'size.value', 'bioassayLabel.value', 'cellLineLabel.value', 'organismLabel.value', 'datasetTitle.value']]

    result_table["mLabel.value"] = result_table["mLabel.value"].str.replace("nanomaterial ","")
    result_table["material_factorized"] = pd.Categorical(result_table["material.value"]).codes
    result_table["material_info"] = "Material " + result_table["material_factorized"].apply(str) + ": " + result_table["mLabel.value"] + " " + result_table["size.value"].astype(str) + " nm"
    
    #result_groups = result_table.groupby(['bioassay.value', 'material.value'])
    result_groups = result_table.groupby(['bioassay.value'])
    
    fig = plt.figure(figsize=(20,30))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.1, hspace=0.6)
    
    rows = math.ceil(float(len(result_groups.groups) / 2))

    subplot_index = -1
    
    for group_i in range(len(result_groups.groups)):
        
        tmp_data = result_groups.get_group(list(result_groups.groups)[group_i])

        if tmp_data["concValue.value"].min() > 1:
            
            subplot_index += 1
            ax = fig.add_subplot(rows, 2, subplot_index + 1)
            ax.set_title("Assay: " + tmp_data.iloc[0,6] + "\n Cell line: " + tmp_data.iloc[0,7] + "\n " + tmp_data.iloc[0,8] +" \n Source: " + tmp_data.iloc[0,9][:40])
    
            ax.set_xlim([0, 105])
            
            g = sns.lineplot(data=tmp_data, x="concValue.value", y="respValue.value", hue="material_info", legend=False)
            g = sns.scatterplot(data=tmp_data, x="concValue.value", y="respValue.value", hue="material_info")

            g.set(ylim=(result_table['respValue.value'].min()-5, result_table['respValue.value'].max()+5), xlabel="concentration mg/ml", ylabel="cell viability %")

            plt.xticks(rotation=25)

    fig.savefig("figures/dose-response.png", dpi=300) 
```

```python
plot_dose_response("npo:NPO_1542")
```

```python
def query_expression_TiO2_venn(GSE):
    """
    Query the endpoint with the given query string and return the results as a pandas Dataframe.
    """
    sparql_query = """

    PREFIX bao: <http://www.bioassayontology.org/bao#>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX dct: <http://purl.org/dc/terms/>
    PREFIX enm: <http://purl.enanomapper.org/onto/>
    PREFIX obo: <http://purl.obolibrary.org/obo/>
    PREFIX npo: <http://purl.bioontology.org/ontology/npo#>

    select distinct ?geneLabel ?avgExpression where { 

        ?bioassay   rdfs:label ?bioassayLabel;
                    dct:identifier """+GSE+""";
                    bao:BAO_0000209 ?mgSampleGroup1.
                            
        ?material   bao:BAO_0003102 bao:BAO_0002932;
                    dct:type obo:CHEBI_51050; # TiO2 type
                    rdfs:label ?mLabel.
        
        ?mgSampleGroup1 a bao:BAO_0000040;
                        rdfs:label ?mgSampleGroupLabel;
                        bao:BAO_0000208 ?timeEndpoint;
                        bao:BAO_0000208 ?concEndpoint;
                        bao:BAO_0090012 ?material.    

        ?timeEndpoint  a bao:BAO_0002114;
                       bao:BAO_0002669 ?timeValue;
                       rdfs:label ?timeLabel.

        ?concEndpoint  a bao:BAO_0000098;
                       bao:BAO_0002666 ?concValue;
                       rdfs:label ?concLabel.
        
        ?mgContrast a bao:BAO_0000040;
                        rdfs:label ?mgContrastLabel;
                        bao:BAO_0095000 ?mgSampleGroup1.

        ?mgDeg a bao:BAO_0000040; # a measure group
                       bao:BAO_0000211 ?gene; # has target
                       bao:BAO_0000208 ?deg; # has endpoint
                       bao:BAO_0000426 ?mgContrast. # is derived from

        ?gene a bao:BAO_0000582;
                rdfs:label ?geneLabel.
          
        ?deg bao:BAO_0000108 ?avgExpression.
         
        FILTER(?concValue = 100.0 && ?timeValue = 24.0)
         
    }
    """
    sparql_service_url = "http://virtuoso:8890/sparql"

    sparql = SPARQLWrapper(sparql_service_url, agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36")  
    
    sparql.setQuery(sparql_query)
    sparql.setReturnFormat(JSON)

    # ask for the result
    result = sparql.query().convert()
    result_table = json_normalize(result["results"]["bindings"])
    
    result_table = result_table.drop_duplicates()
    
    return result_table
```

```python
res1 = query_expression_TiO2_venn('"GSE42068"@en')
res2 = query_expression_TiO2_venn('"GSE42067"@en')
```

```python
res1.to_csv("data/res1.csv", index=False, encoding="utf-8")
res2.to_csv("data/res2.csv", index=False, encoding="utf-8")
```
