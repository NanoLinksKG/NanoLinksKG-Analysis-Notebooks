```python
import pandas as pd
```

## Dataset 01

```python
df = pd.read_csv("datasets/dataset-01/original-dataset.csv", encoding="utf-8")
```

```python
df.shape
```

```python
df.rename(columns={'surf_area': 'specific_surf_area'}, inplace=True)
```

```python
df.head()
```

```python
df.insert(loc=0, column='rownum', value=df.index+1)
```

```python
df.head()
```

```python
df.insert(loc=1, column='bioassay_id', value=df.set_index(['assay','cell_line', 'cell_species', 'cell_origin', 'cell_type']).index.factorize()[0]+1)
```

```python
df.head()
```

```python
df.bioassay_id.unique()
```

```python
df.insert(loc=2, column='material_id', value=df.set_index(['material','core_size', 'hydro_size', 'surf_charge', 'specific_surf_area']).index.factorize()[0]+1)
```

```python
df.head()
```

```python
df.material_id.unique()
```

```python
len(set(df.set_index(['cell_line','cell_species', 'cell_origin', 'cell_type']).index.factorize()[0]+1))
```

```python
len(df.cell_line.unique())
```

```python
df.to_csv("datasets/dataset-01/data.csv", index=False, encoding="utf-8")
```

## Dataset 05

```python
df = pd.read_csv("datasets/dataset-05/original-dataset.csv", encoding='utf-8')
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop('row_num', axis=1, inplace=True)
```

```python
df.insert(loc=0, column='rownum', value=df.index+1)
```

```python
df.head()
```

```python
df.animal_species.unique()
```

```python
df.loc[(df.human_or_animal == "H") & (df.animal_species.isna()), 'animal_species'] = "Human"
df.loc[(df.human_or_animal == "A") & (df.animal_species.isna()), 'animal_species'] = "unknown"
```

```python
df.animal_species.unique()
```

```python
df.insert(loc=1, column='bioassay_id', value=df.set_index(['reference_doi','test_assay','cell_type', 'animal_species', 'cell_organ_or_tissue', 'cell_line_or_primary']).index.factorize()[0]+1)
```

```python
df.head()
```

```python
df.bioassay_id.unique()
```

```python
df.insert(loc=2, column='material_id', value=df.set_index(['reference_doi','nanoparticle','diameter', 'zeta_potential', 'coat', 'type_organic_inorganic']).index.factorize()[0]+1)
```

```python
df.head()
```

```python
df.material_id.unique()
```

```python
df.insert(loc=3, column='cell_type_id', value=df.set_index(['reference_doi','cell_type', 'animal_species', 'human_or_animal', 'cell_morphology','cell_organ_or_tissue','cell_line_or_primary']).index.factorize()[0]+1)
```

```python
df.head()
```

```python
len(df.cell_type_id.unique())
```

```python
len(set(df.set_index(['cell_type', 'animal_species', 'cell_organ_or_tissue', 'cell_line_or_primary']).index.factorize()[0]+1))
```

```python
len(set(df.set_index(['cell_type', 'animal_species', 'human_or_animal', 'cell_morphology','cell_organ_or_tissue','cell_line_or_primary']).index.factorize()[0]+1))
```

```python
len(df.cell_type.unique())
```

```python
df.to_csv("datasets/dataset-05/data.csv", index=False, encoding="utf-8")
```
